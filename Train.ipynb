```python
from ultralytics import YOLO
```

```python
model=YOLO("yolov8n.pt")
```

##Traffic (car + truck):

```python
model.train(data ="/content/dataset.yaml", epochs=50, imgsz=640)
```

```python
import cv2
from google.colab.patches import cv2_imshow
cfs=cv2.imread('/content/drive/MyDrive/KLTN/traffic/runs/detect/train/confusion_matrix.png')
cv2_imshow(cfs)
```

```python
results=cv2.imread('/content/drive/MyDrive/KLTN/traffic/runs/detect/train/results.png')
cv2_imshow(results)
```

```python
model=YOLO("/content/drive/MyDrive/KLTN/traffic/runs/detect/train/weights/best.pt")
model.val(data="/content/dataset.yaml", split="test")
```
